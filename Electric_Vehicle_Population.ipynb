```python
range_electric = df.groupby("Make")[["Electric Range"]].mean().reset_index()
plt.figure(figsize=(18,5))
plt.bar(x="Make",height="Electric Range",data=range_electric)
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(18,5))
sns.lineplot(x="Electric Vehicle Type",y="Electric Range",hue="Model Year",data=df)
plt.xticks(rotation=90);
```
