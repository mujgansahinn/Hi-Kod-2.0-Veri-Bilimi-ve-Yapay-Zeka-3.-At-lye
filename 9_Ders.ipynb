```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
date_method_sold = m.groupby(["Invoice Date","Sales Method"])[["Units Sold"]].sum().reset_index()
date_method_sold
```

```python
in_store = date_method_sold[date_method_sold["Sales Method"]=="In-store"]
online = date_method_sold[date_method_sold["Sales Method"]=="Online"]
outlet = date_method_sold[date_method_sold["Sales Method"]=="Outlet"]
```

```python
in_store
```

```python
fig = make_subplots(rows = 3, cols= 1)

fig.append_trace(go.Line(x = in_store["Invoice Date"], y = in_store["Units Sold"]), row = 1, col = 1)

fig.append_trace(go.Line(x = online["Invoice Date"], y = online["Units Sold"]), row = 2, col = 1)

fig.append_trace(go.Line(x = outlet["Invoice Date"], y = outlet["Units Sold"]), row = 3, col = 1)

fig.show()
```

```python
fig = make_subplots(rows = 3, cols= 1)

fig.append_trace(go.Line(x = in_store["Invoice Date"], y = in_store["Units Sold"]), row = 1, col = 1)

fig.append_trace(go.Line(x = online["Invoice Date"], y = online["Units Sold"]), row = 2, col = 1)

fig.append_trace(go.Line(x = outlet["Invoice Date"], y = outlet["Units Sold"]), row = 3, col = 1)

fig.show()
```
